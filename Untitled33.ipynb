```python
import os
import torch
import torch.nn as nn
import torchvision
from video_diffusion_pytorch import Unet3D, GaussianDiffusion
import matplotlib.pyplot as plt
from diffusers import AutoencoderKL,DDPMScheduler
from tqdm import tqdm
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
import torchvision.transforms.functional as F

def preprocess(video_path, min_frames, target_size=(256, 256),config =None,batch_size=8,device=None):

  video, _, info = torchvision.io.read_video(video_path)
  num_frames = video.shape[0]

  if num_frames > min_frames:
    video = video[:min_frames, :, :, :]
  elif num_frames < min_frames:
        pad_frames = min_frames - num_frames
        padding = torch.zeros(
            (pad_frames, *video.shape[1:]),
            dtype=video.dtype
          )
        video = torch.cat([video, padding], dim=0)


  resized_video = torch.stack([F.resize(frame.permute(2, 0, 1), target_size).permute(1, 2, 0) for frame in video])
  resized_video = (resized_video.float() / 255 - 0.5) * 2

  video_tensor=resized_video.permute(0,3,1,2)
  ref_frame = resized_video[0].permute(2, 0, 1)


  return video_tensor,ref_frame
```

```python
video_path='/content/image_to_video/dataset/UCF101/UCF-101'
list_video=os.listdir(video_path)
data=[]
for video_directory_name in list_video:
  video_directory_path = os.path.join(video_path, video_directory_name)
  if os.path.isdir(video_directory_path):
    for video_file_name in os.listdir(video_directory_path):
      data.append(os.path.join(video_directory_path, video_file_name))

random.shuffle(data)

vae = AutoencoderKL.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    subfolder="vae"
  ).to(device)
dataset = VideoDataset(data, vae, device=device, min_frames=30)
dataloader = DataLoader(dataset, batch_size=4, shuffle=True)
```

```python
def save_checkpoint(model, optimizer,epoch,checkpoint_path="content/drive/checkpoint.pth"):
    torch.save({
        "epoch": epoch,
        "state_dict": model.state_dict(),
        "optimizer": optimizer.state_dict(),
      },checkpoint_path)
    print(f"Checkpoint saved at epoch {epoch}.")
```

```python
def load_checkpoint(model, optimizer,checkpoint_path="content/drive/checkpoint.pth"):
   if os.path.exists(checkpoint_path):
    checkpoint=torch.load(checkpoint_path)
    model.load_state_dict(checkpoint['model_state_dict'])
    optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
    start_epoch = checkpoint['epoch'] + 1
    print(f"Resumed from checkpoint at epoch {checkpoint['epoch']}.")

   else:
    start_epoch=0
    print("No checkpoint found.")
   return start_epoch
```

```python
import torchvision.transforms.functional as F

def training_step(model, noisy_latents, t, noise, ref_images, optimizer):
  predicted_noise=model(noisy_latents,t,ref_images)
  loss=nn.MSELoss()(predicted_noise,noise)
  loss.backward()
  optimizer.step()
  return loss
```

```python
def training_loop(dataloader,epoch,batch_size,device,list_video):
  scheduler=DDPMScheduler(num_train_timesteps=1000)
  model=Unet3D(dim=64,dim_mults=(1, 2, 4),channels=4).to(device)
  optimizer = torch.optim.Adam(model.parameters(), lr=1e-6)
  start_epoch=load_checkpoint(model,optimizer)
  for e in range(start_epoch,epoch):
        print(f"Epoch {e} started...")
        pbar=tqdm(total=len(dataloader))
        for latent,ref_images in dataloader:
          latent=latent.to(device)
          ref_images = ref_images.unsqueeze(2)
          ref_images = ref_images.expand(-1, -1, T, -1, -1)
          ref_images=ref_images.to(device)
          t=torch.randint(0,scheduler.num_train_timesteps,(batch_size,),device=device)
          latent=latent.to(device)
          ref_images=ref_images.to(device)
          noise=torch.randn_like(latent).to(device)
          noisy_latents=scheduler.add_noise(latent,noise,t).to(device)

          optimizer.zero_grad()
          loss=training_step(model, noisy_latents, t, noise, ref_images, optimizer)
          pbar.set_postfix({"loss": loss.item()})
          pbar.update(1)
        save_checkpoint(model, optimizer, e)
```
